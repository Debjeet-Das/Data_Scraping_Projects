## Parsing the Source Code :
  - Here we are importing the bs4 library which will help us to scrape the data.
  - Now the objective is to make the source code of the webpage parsable.
  - Making it parsable will help us to fetch the desired data and can be further used for analytics.
  - After that we will make a soup component which will contain the parsable source code of the web page.

```python
# importing bs4
import bs4

# making a soup component which will contain the parsable html source code of the web page to be scrapped
soup=bs4.BeautifulSoup(html,"html.parser")
soup
```

## Scrapping Data :
   - The first step is to go to the browser from where we are scrapping the data and look at its source code.
   - Then we need to inspect the elements we want to fetch from the source code.
   - Now the source code cosists of html tags, class and attributes which we are concerned about for scrapping.
   - Each element will reside inside a parent tag or children of that parrent tag.
   - There will be class defined for the elements to be fetched also with id which will be unique for all, along with some attributes.
   - So now we will scrape the data which has information about the various data science jobs available and some other parameters related to it.


## Elements to be scrapped :
   - We will scrape the designations offered by different companies.
   - Job location.
   - Minimum Expirience required for them.
   - Salary offered.
   - Name of the Company offering the job.

```python
# extracting designation which resides inside the tag "li" and class "desig"

# we are creating an empty list in order to append all the designation to it.
designation=[]

# we are using the find_all function to find all the tags "li" and class_= "desig" from the parsed source code.
for i in soup.find_all("li",class_="desig"):
    designation.append(i.text.strip())


# So in the above step we are fetching the designations from the source code and then appending it to the list of designations.
designation    
```

```python
# extracting locations which resides inside the tag "span" and class "loc"

location =[]
for i in soup.find_all("span",class_="loc"):
    location.append(i.text.strip())

# this list will contain all the job locations 
location
```

```python
# extracting experience which is p[resent inside the tag "span" and class "exp"
exp= []
for i in soup.find_all("span",class_="exp"):
    exp.append(i.text.split("-")[0].strip())


# this list will contain the minimum experience requirement by the companies offering the job.
exp
```

```python
# extracting salary offered by the companies
sal= []

# Salary offered is present inside the tag "span" and class "salary"
for i in soup.find_all("span",class_="salary"):
    sal.append(i.text.strip())

    
# this loist will contain the salaries offered by the companies for a specific job designation    
sal
```

```python
#extracting company name or the organisation name offering the job
org= []
for i in soup.find_all("span",class_="org"):
    org.append(i.text.strip().upper())
    

    
# List containing all the Organisation names offering jobs.    
org
```

## Creating a dataframe :
   - Now we will create a dataframe from the extracted data which can be further used for analytics.
   - This dataframe consists of Company name, Salary offered, Job location and designation. 

```python
import pandas as pd
data2= pd.DataFrame({"company": org,"salary":sal,"experience":exp,"location":location,"designation":designation})
```

## Making a user defined function :
   - It will have one argument which will be the ural of the web page to be scrapped.
   - This function will extract all the information about a particular job opening when called.
   - It will open a browser, start fetching the source code and making it parsable.
   - Then will extract the information about the job openings  offered by the different companies.
   - Which will return a dataframe containing all the above information.

```python
# defining functions to get salary , designation, location,compname,exp

def get_jos(x):
    
    from selenium import webdriver
    import os

    path_to_unzipfile= "C:\\Users\\Debjeet Das\\Desktop\\PYTHON"
    os.chdir(path_to_unzipfile)
    browser1= webdriver.Firefox(path_to_unzipfile)
    
    browser1.get(x)
    html=browser1.page_source
    
    import bs4
    soup=bs4.BeautifulSoup(html,"html.parser")
    
    location =[]
    for i in soup.find_all("span",class_="loc"):
        location.append(i.text.strip())
    
    designation=[]
    for i in soup.find_all("li",class_="desig"):
        designation.append(i.text.strip())
        
    exp= []
    for i in soup.find_all("span",class_="exp"):
        exp.append(i.text.split("-")[0].strip())
        
    
    sal= []
    for i in soup.find_all("span",class_="salary"):
        sal.append(i.text.strip())
        
    org= []
    for i in soup.find_all("span",class_="org"):
        org.append(i.text.strip().upper())
        
    import pandas as pd
    data2= pd.DataFrame({"company": org,"salary":sal,"experience":exp,"location":location,"designation":designation})
    
    return data2
```

```python
# calling the user defined fuction and returning the resultant dataframe with necessary informations scrapped from the web page.
url = "https://www.naukri.com/data-science-jobs"
get_jos(url)
```

## Selenium advantages :
   - Using selenium we can depict a full fledged web browser.
   - This will help us to extract data from many pages of the same website.
   - So using various fuctions of selenium we can hover through all the pages and extract data from it.
   - Thus using selenium we can click on the buttons on a particular website to perform certain action from remote user.
   - In the below code we are finding elements by class name which consists of button, and the clicking on the button using .click()
   - In this way we can build a full fledged web browser and control it form remote user using selenium.

```python
url1 = "https://www.naukri.com/data-science-jobs"
browser.get(url1)


# btns1 will contain a list of all the buttons which is present in the web page.
btns1=browser.find_elements_by_class_name("grayBtn")

btns1
```

## Creating Final Dataframe :
   -  This dataframe will contain all the informations scrapped using the above user defined function.
   - In the below code we are first creating an empty dictionary with key and value pair.
   - Keys will be the page numbers and values will be the data frame corresponding to the particular page.
   - We are using the .click() function to hover through the pages and extract data.
   - so the resultant dictionary will consist of many dataframes with desired information scrapped from all the pages.

```python
url2 = "https://www.naukri.com/data-science-jobs-{}"

main={}

for i in range(1,5):
    main[i]=get_jos(url2.format(i))
    
    btns2=browser.find_elements_by_class_name("grayBtn")
    for i in btns2:
        if i== "NEXT":
             btns2.click()
    
```

```python
# This is the datframe created from the data extracted from the 1st page of tghe naukri website,stored in the ductionary
main[1]
```

```python
# Here we are concating all the dataframes present in the above formed dictionary for further analysis. 

import pandas as pd
big_data1= pd.concat([main[i] for i in main],axis =0)
```

```python
# Reseting the index of the above formed final data frame consisting of information stored in it extracted from all the pages.

big_data1=big_data1.reset_index()
big_data1.drop("index",inplace=True,axis=1)
```
