```python
# importing libraries

import requests
import json


# url from where the data is to be scrapped
url="https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2016-10-01&endtime=2016-10-02"

# this will extract the source code of the given url and store it in text format in request variable
request=requests.get(url)
```

```python
# we are converting the source code tojson format
json_data=request.json()
```

```python
# this is how it looks after conversion
json_data
```

```python
# we are extracting the signal from the above file
signal=[] # creating an empty list
for i in json_data["features"]:   # here we are iterating through all the feature keys present in the json file
    signal.append(i["properties"]["sig"]) # appending it to the signal list
    
signal
```

```python
# keys in the json_data
for i in json_data:
    print(i)
```

```python
# Extracting magnitude
magnitude=[]
for i in json_data["features"]:
    magnitude.append(i["properties"]["mag"])
magnitude
```

```python
# extracting urls
url=[]
for i in json_data["features"]:
    url.append(i["properties"]["url"])
url
```

```python
# Extracting latitudes
latitude=[]
for i in json_data["features"]:
    latitude.append(i["geometry"]["coordinates"][0])
latitude
```

```python
# Extracting longitude
longitude=[]
for i in json_data["features"]:
    longitude.append(i["geometry"]["coordinates"][1])
longitude    
```

```python
# Extracting Places
title=[]
for i in json_data["features"]:
    title.append(i["properties"]["title"])
title
```

```python
# Extracting Place
place=[]
for i in json_data["features"]:
    place.append(i["properties"]["place"])
place
```

```python
# Creating data frame
import pandas as pd
eq_data= pd.DataFrame({"Place":place,"Title":title,"Longitude":longitude,"Latitude":latitude,"Magnitude":magnitude})
```

```python
# no. of occasions when the mag was > 2
len(eq_data[eq_data["Magnitude"]>2])
```

```python
# converting the file to a csv file so as to use for further analysis
eq_csv=eq_data.to_csv("C:\\Users\\debjeet das\\Desktop\\PYTHON\\earthquakecsv.csv")
```
